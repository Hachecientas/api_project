```python
import requests
import pandas as pd

url_all = 'http://127.0.0.1:5000/api/v1/dundermifflin/datos/all'
url_train ='http://127.0.0.1:5000/api/v1/dundermifflin/datos/train'
url_test = 'http://127.0.0.1:5000/api/v1/dundermifflin/datos/test'
url_preds = 'http://127.0.0.1:5000/api/v1/dundermifflin/datos/predicciones'
url_query = 'http://127.0.0.1:5000/api/v1/dundermifflin/datos?year=2011'
url_28905 = 'http://127.0.0.1:5000/api/v1/dundermifflin/datos/28905'
```

```python
r_all = requests.get(url_all)
r_all.json()
```

```python
r_train = requests.get(url_train)
r_train.json()
```

```python
r_test = requests.get(url_test)
r_test.json()
```

```python
r_preds = requests.get(url_preds)
r_preds.json()
```

```python
r_28905 = requests.get(url_28905)
r_28905.json()
```

```python
r_query = requests.get(url_query)
r_query.json()
```

```python
req = [r_all,r_train,r_test,r_preds,r_28905,r_query]

responses = [r.status_code for r in req]

responses
```
