```python

@app.route('/', methods=['GET'])
def home():
    image_url = url_for('static', filename='dunder_mifflin_logo.jpg')
    return web

@app.route('/api/v1/dundermifflin/datos/all', methods=['GET']) 
def api_all():
    return jsonify(datos['ventas2'])


@app.route('/api/v1/dundermifflin/datos/train', methods=['GET']) 
def api_train():
    return jsonify(datos['X_train'])


@app.route('/api/v1/dundermifflin/datos/test', methods=['GET']) 
def api_test():
    return jsonify(datos['X_test'])


@app.route('/api/v1/dundermifflin/datos/predicciones', methods=['GET']) 
def api_preds():
    return jsonify(datos['preds_test_df'])

@app.route('/api/v1/dundermifflin/datos/28905', methods=['GET']) 
def api_28905():
    return jsonify(pd.DataFrame(datos['ventas2']).iloc[28905].to_dict())

@app.route('/api/v1/dundermifflin/datos', methods=['GET'])
def api_year():

    if 'year' in request.args:
        year = int(request.args['year'])
    else:
        return "Error: No se ha especificado un ao. Por favor introduce uno"

    data = [x for x in datos['ventas2'] if x['year'] == year]

    return jsonify(data)
```
